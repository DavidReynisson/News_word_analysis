# Step 5: Simple Linear Regression

Here we regress headline upvotes and number of comments on their sentiments. 

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('mlr_titles.csv').drop(['Unnamed: 0'], axis=1)
```

```python
#split into two dataframes, negative and positive
df_positive = df[df.sentiment > 0]
df_negative = df[df.sentiment < 0]
```

```python
import statsmodels.api as sm
 
# defining the variables
x = df_positive['sentiment'].tolist()
y = df_positive['score'].tolist()
y2 = df_positive['num_comments'].tolist()
 
      
# adding the constant term
x = sm.add_constant(x)
 
# performing the regression
# and fitting the model
result = sm.OLS(y, x).fit()
 
# printing the summary table
print(result.summary())

# now for comments
result1 = sm.OLS(y2, x).fit()
print(result1.summary())
```

```python
import statsmodels.api as sm
 
# defining the variables
x = df_negative['sentiment'].tolist()
y = df_negative['score'].tolist()
y2 = df_negative['num_comments'].tolist()
 
      
# adding the constant term
x = sm.add_constant(x)
 
# performing the regression
# and fitting the model
result = sm.OLS(y, x).fit()
 
# printing the summary table
print(result.summary())

# now for comments
result1 = sm.OLS(y2, x).fit()
print(result1.summary())
```
